```python
import matplotlib.pyplot as plt

squares = [1, 4, 9, 16, 25]
plt.plot(squares)
plt.show()
```

```python
import matplotlib.pyplot as plt
squares = [1, 4, 9, 16, 25]
plt.plot(squares, linewidth=5)
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.tick_params(axis='both', labelsize=24)
plt.show()
```

```python
# single scattered point
import matplotlib.pyplot as plt
plt.scatter(2, 4, s = 200) 

# set title and lables
plt.title("Square Numbers", fontsize=24) 
plt.xlabel("Value", fontsize=14) 
plt.ylabel("Square of Value", fontsize=14)

# set param size
plt.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```

```python
# multiple scattered points
import matplotlib.pyplot as plt
x_values = [1, 2, 3, 4, 5] 
y_values = [1, 4, 9, 16, 25]
# remove edge color in case there are too many points
plt.scatter(x_values, y_values, edgecolor='none', s=40)

# set title and lables
plt.title("Square Numbers", fontsize=24) 
plt.xlabel("Value", fontsize=14) 
plt.ylabel("Square of Value", fontsize=14)

# set param size
plt.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```

```python
# change line color and colormap
import matplotlib.pyplot as plt
x_values = list(range(1001)) 
y_values = [x**2 for x in x_values]
# plt.scatter(x_values, y_values, c='pink',edgecolor='none', s=40) 
plt.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, 
            edgecolor='none', s=40) # COLOR MAP

# set title and lables
plt.title("Square Numbers", fontsize=24) 
plt.xlabel("Value", fontsize=14) 
plt.ylabel("Square of Value", fontsize=14)


plt.axis([0, 1100, 0, 1100000]) 
plt.show()
```

```python
# save image
import matplotlib.pyplot as plt
x_values = list(range(1001)) 
y_values = [x**2 for x in x_values]
plt.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, 
            edgecolor='none', s=40) # COLOR MAP

# set title and lables
plt.title("Square Numbers", fontsize=24) 
plt.xlabel("Value", fontsize=14) 
plt.ylabel("Square of Value", fontsize=14)


plt.axis([0, 1100, 0, 1100000]) 
plt.savefig('squares_plot.png', bbox_inches='tight')
```

```python
# exercise 15-1 & 15-2 
# Lesson learned 
#     - axis only take list data type
#     - x_values, y_value, scatter(), title, xlabel, ylabel, axis
import matplotlib.pyplot as plt
x_values = list(range(1,5000))
y_values = list(x**3 for x in range(1,5000))
plt.scatter(x_values, y_values, c=y_values, cmap = plt.cm.Blues,
            edgecolor = 'none', s = 40)

# set titles and lablee
plt.title('3rd Power Numbers',fontsize = 24)
plt.xlabel('Value', fontsize = 14)
plt.ylabel('3rd Power of Value', fontsize = 14)

plt.axis([0, 5000, 0, 100000000000]) 
plt.show()
```
